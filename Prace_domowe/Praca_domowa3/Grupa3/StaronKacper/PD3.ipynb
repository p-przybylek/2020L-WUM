## Przygotowanie danych i środowiska

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('../../australia.csv')
```

```python
data
```

## Klasyfikatory

```python
tree = DecisionTreeClassifier(max_depth = 200, random_state = 1)
kn = KNeighborsClassifier(n_neighbors = 10, algorithm = 'kd_tree')
reg = LogisticRegression(random_state = 1, max_iter=1000, penalty = 'l2')

scores = []
```

### Podział na zbiór treningowy i testowy

```python
X = data.drop(['RainTomorrow'], axis=1)
y = data['RainTomorrow']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

### 1. DecisionTreeClassifier

```python
tree.fit(X_train, y_train)
y_pred = tree.predict(X_test)

scores.append([accuracy_score(y_pred, y_test), 
              precision_score(y_pred, y_test), 
              recall_score(y_pred, y_test),
              f1_score(y_pred, y_test),
              roc_auc_score(y_pred, y_test)])
```

### 2. KNeighboursClassifier

```python
kn.fit(X_train, y_train)
y_pred = kn.predict(X_test)

scores.append([accuracy_score(y_pred, y_test), 
              precision_score(y_pred, y_test), 
              recall_score(y_pred, y_test),
              f1_score(y_pred, y_test),
              roc_auc_score(y_pred, y_test)])
```

### 3. LogisticRegression

```python
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)

scores.append([accuracy_score(y_pred, y_test), 
              precision_score(y_pred, y_test), 
              recall_score(y_pred, y_test),
              f1_score(y_pred, y_test),
              roc_auc_score(y_pred, y_test)])
```

## Ocena działania

```python
pd.DataFrame(scores, columns = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc'],
             index = ['DecisionTree', 'KNeighbours', 'LogisticRegression'])
```

Biorąc pod uwagę wartości policzone dla poszczególnych miar dla poszczególnych modeli, bez strojenia na zadanym zbiorze najlepiej ze sprawdzonych algorytmów klasyfikacji radzi sobie LogisticRegression
