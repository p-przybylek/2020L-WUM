## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.decomposition import FastICA
```

## Read data

```python
data = pd.read_csv('../../signals.tsv', index_col=0, sep='\t')
```

## Data info

```python
data
```

```python
data.info()
```

## Function

```python
def plot_data(data):
    fig, ax = plt.subplots(2, 2, figsize=(10, 4))
    ax[0, 0].plot(data.index, data.iloc[:, 0])
    ax[0, 1].plot(data.index, data.iloc[:, 1])
    ax[1, 0].plot(data.index, data.iloc[:, 2])
    ax[1, 1].plot(data.index, data.iloc[:, 3])
```

## Plot data

```python
plot_data(data)
```

## ICA - Independent Component Analysis

```python
ica = FastICA(n_components=4, random_state=123)
out = ica.fit_transform(data)
signals = pd.DataFrame(out)
```

```python
plot_data(signals)
```
